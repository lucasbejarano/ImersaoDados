```python
#Solução para uma dúvida em Discord - desafio 07
import matplotlib.pyplot as plt

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title("Gráfico de barras para tratamentos", fontweight="bold", color="#2874A6",fontsize=20)
ax.set_xlabel("Tratamento", fontweight="bold", fontsize="large", color="#566573")
ax.set_ylabel("Quantidade", fontweight="bold", color = "#566573", fontsize="large")


dados_trat = dados['tratamento'].value_counts()

a = dados_trat.loc['com_droga']
b = dados_trat.loc['com_controle']

ax.bar("com droga",a, label="Com droga", color="green")
ax.bar("com controle", b, label="Com controle", color="red") 

plt.legend()
```

```python
#Desafio 07: Deixar os graficos bonitoes. (Matplotlib).

fig,ax= plt.subplots()
fig.patch.set_facecolor("#CBCDCB")


labels=['Com droga','Com tratamento']
colors=['#C44B4F','#607D86']
ax.pie(dados['tratamento'].value_counts(),labels=labels,explode=[0,0.3], colors=colors,autopct='%1.1f%%')
ax.set_title("Tratamentos",fontsize=15)
```

```python
#Desafio 07: Deixar os graficos bonitoes. (Matplotlib).

fig,ax= plt.subplots()
fig.patch.set_facecolor("#BED7D8")

x = ['24','48','72']
colors=['#E151AF','#2BB8B3','#FED43F']
ax.bar(x,dados['tempo'].value_counts().sort_index(),width=0.8,color=colors)
ax.set_title("Tempo Experimentos",fontsize=15,color='white')
plt.xlabel("Quantidade horas")
plt.ylabel("Num Experimentos")
```
